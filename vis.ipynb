```python
'''
munges the pois of Copenhagen such a museums and churches and parks and so on
'''

import folium
import json
import pandas as pd
import geohash
import time
from shapely.geometry import shape, Point, Polygon
from shapely.prepared import prep # speed up


df = pd.DataFrame.from_csv('data/stednavn_flade.csv')
# Polygon(

def poligonify(row):
    a = row.wkb_geometry
    try:
        c =  np.array( [float(i.rstrip(',')) for i in a[10:].translate(None, '()').split(' ')])
        c = c.reshape(c.shape[0]/2, c.shape[0]/(c.shape[0]/2))
        Polygon(c)
    except:
        print 'error'
    return Polygon(c)
```

```python
df['poly']  = df.apply(poligonify, axis=1)
```

```python
df['centroids'] = df.apply(lambda row: Point(row.poly.centroid), axis=1)
```

Guen traffic data

```python
trafic = pd.io.json.read_json('data/trafiktal.json')
trafic['points'] = trafic.apply(lambda row: Point(row.coords), axis=1)
```

```python
# load GeoJSON file containing sectors
with open('data/taxzone.json', 'r') as f:
    js = json.load(f)
```

```python
polygons_shape = [shape(feature['geometry']) for feature in  js['features']]
names = [feature['properties']['id'] for feature in js['features']]
map_df =  pd.DataFrame({'poly':polygons_shape,'id':names})
```

```python
def do(row):
    return len(filter(row.poly.contains, df.centroids))
```

```python
res = {}
for polygon in map_df.poly:
    counts =0 
    for point in trafic.points:
        if polygon.contains(point):
            counts += int(trafic['count'][trafic.points==point])
    res[polygon] = counts
```

```python
map_df_trafic = pd.DataFrame({'poly':res.keys(),'counts':res.values()})
map_df_trafic.head()
```

```python
map_df['poicounts'] = map_df.apply(do,axis=1)
```

```python
map_df.head()
```

```python
data = pd.concat([map_df,map_df_trafic])
```

```python
data['index'] = 2*data.poicounts + 0.5 * data.counts
```

```python
cph_location = [56.3,11]
outfile = 'cph.html'
geo_path = 'data/taxzone.json'
map = folium.Map(location=cph_location, zoom_start=7, tiles='Stamen Toner',width=700, height = 700)
map.geo_json(geo_path=geo_path, data_out ='test.json', data=map_df_trafic,
             columns=['id', 'counts'],
             key_on='feature.properties.id',
             fill_color='YlGnBu', fill_opacity=0.8, line_opacity=0.3,
             legend_name='trafficcounts')
map.create_map(path=outfile)
```
